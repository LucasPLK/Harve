```python
import pandas as pd
```

```python
dffifa = pd.read_csv('https://www.harve.com.br/praticas/fifaplayers_pt.csv')
dffifa.head(2)
```

```python
dffifa.columns
```

```python
# Frequência absoluta
dffifa['posição'].value_counts()
```

```python
# Frequência relativa:
frequencia_absoluta = dffifa['posição'].value_counts()
total = frequencia_absoluta.sum()
frequencia_relativa = frequencia_absoluta/total

frequencia_relativa
```

```python
frequencia_absoluta.plot(kind='barh');
```

```python
dados = {'id_familia':[1,2,3,4,5,6,7,8],
         'tipo_casa': ["apartamento","casa","casa","kitnet","sobrado","apartamento","apartamento","casa"],
         'tamanho_casa': ["pequeno","médio","grande","pequeno","médio","grande","grande","pequeno"],
         'n_cartoes': [4,6,6,7,8,7,8,10],
         'tamanho_familia': [2,2,4,4,5,5,6,6],
         'renda_familiar':[14,16,14,17,18,21,17,25],
         'num_automoveis': [1,2,2,1,3,2,1,3],
         'distanciacentro': [22.2,3,31,2,1,12,2,4]}#km

df = pd.DataFrame(dados)
df.head()
```

```python
df.groupby('tamanho_casa').count()[['n_cartoes']]
```

```python
df.groupby('tamanho_casa').count()[['n_cartoes']]/df.shape[0]
```

```python
df.groupby('tamanho_casa').count()[['n_cartoes']].plot(kind='bar')
```

# Tabela de contingências

```python
# dffifa['posição']
# dffifa['perna de preferência']
tabela_contingencias_fifa = pd.crosstab(dffifa['posição'], dffifa['perna de preferência'])
tabela_contingencias_fifa
```

```python
# Frequência relativa:
tabela_contingencias_fifa*100/dffifa.shape[0]
```

```python
tabela_contingencias_fifa.plot(kind='bar');
```

```python
contingengecia_casas = pd.crosstab(df['tipo_casa'], df['tamanho_casa'])
contingengecia_casas
```

```python
contingengecia_casas.T.plot(kind='bar');
```
