# Arvores de decisões

Vamos identificar qual a mensuração de diabetes de cada paciente e predizer um paciente é diabético ou não com base nas features apresentadas.

```python
from sklearn.datasets import load_diabetes #dataset para casos de diabetes
from sklearn.tree import DecisionTreeClassifier, plot_tree

from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split # separação do dataset para validação e treinamento
from sklearn.metrics import classification_report # métricas de avaliação do modelo

import seaborn as sns # biblioteca par geração de gráficos
import matplotlib.pyplot as plt
```

# Carregamento e preparação de dados

```python
dados = load_diabetes(as_frame=True)
```

```python
X = dados['data']
Y = (dados['target'] > 100).astype(int) # vamos considerar esse dataset

x_train, x_test, y_train, y_test = train_test_split(X, Y)
```

# Modelo de árvore de decisão

```python
modelo = DecisionTreeClassifier()
modelo.fit(x_train, y_train)
```

# Verificando as decisões:
Vamos verificar as 3 primeiras decisões que a árvore aprendeu

```python
plt.rcParams['figure.figsize'] = (12,10) # vamos alterar o tamanho da figura padrão do matplotlib para conseguirmos ver os dados com mais clareza
plot_tree(modelo, max_depth=3, fontsize=8, feature_names=dados.feature_names);
# vamos usar o ";" (ponto e virgula) para evitar uma grande quantidade de outputs textuais no console e termos unicamente o plot.
```

```python
y_pred = modelo.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

# Randon Forest

```python
modelo_rf = RandomForestClassifier()
modelo_rf.fit(x_train, y_train)
```

```python
y_pred_rf = modelo_rf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_rf))
```
